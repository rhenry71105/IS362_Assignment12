<h><b> We begin by importing the necessary libraries. </b></h>

```python
import pandas as pd
import json, requests
from pandas.io.json import json_normalize
```

<h><b> We proceed by requesting the API data. </b></h>

```python
#The data is obtained from - http://developer.nytimes.com/movie_reviews_v2.json#/Console/GET/reviews/search.json
#Once signed up NYT provides an API Key: 'bkJtO2rcSlxIy2sd1JzHSdASkGDnxAGI'

api_key = 'bkJtO2rcSlxIy2sd1JzHSdASkGDnxAGI'
url = 'https://api.nytimes.com/svc/movies/v2/reviews/search.json?'
string = url + '&api-key=' + api_key

repos = json.loads(requests.get(string).text) #getting data to display
repos
```

<h><b> For easier viewing we can proceed by normalizing the data. </b></h>

```python
normal = json_normalize(repos['results'])
normal
```

<h><b> Now we can display the data by three columns (byline, display_title, & summary_short). This will allow viewers to see the reviews easier. </b></h>

```python
df = normal[['byline', 'display_title', 'summary_short']]
df
```
